```python
edges = {('','A',3), ('H','B',2), ('H','C',5),
 ('A','D',3),
 ('B','D',1), ('B','E',6),
 ('C','E',2),
 ('D','B',1), ('D','F',4),
 ('E','F',1), ('E','S',4),
 ('F','E',1), ('F','S',2)}

edges
```

```python
G = nx.DiGraph()
G
```

```python
G.add_weighted_edges_from(edges)
G.nodes
```

```python

print(sorted(G.edges))
```

```python
pos = nx.spring_layout(G, scale = 1000.)
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
```

```python
b = 'H'
f = 'S'
```

```python
paths = nx.all_simple_paths(G, b, f)
```

```python
for path in list(paths):
 print(path)
```

```python

help(nx.dijkstra_path)
#dijkstra_path(G, source, target, weight='weight')
#Returns the shortest weighted path from source to target in G.
```

```python

path = nx.dijkstra_path(G, b, f)
path
```

```python
path_length = nx.dijkstra_path_length(G, b, f)
path_length
```

```python

print("The shortest path from", b, "to", f, "is: ", path, "with length", path_length)
```
