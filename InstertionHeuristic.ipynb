```python
def insertion_heuristic(d, nodes, tour):
    while len(tour) < len(nodes):
        min_insertion_cost = float('inf')  # initialize with infinity
        for i in nodes:
            if i not in tour:
                # Calculate the insertion cost for each node not in the tour
                for j in range(len(tour)-1):
                    cost = d[tour[j]][i] + d[i][tour[j+1]] - d[tour[j]][tour[j+1]]
                    if cost < min_insertion_cost:
                        min_insertion_cost = cost
                        node_to_insert = i
                        insert_position = j+1
        # Insert the node at the position where the cost is minimum
        tour.insert(insert_position, node_to_insert)
        print('Inserted', node_to_insert, 'into the tour!')
    return tour
```
