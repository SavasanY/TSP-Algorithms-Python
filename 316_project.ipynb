```python
df = pandas.read_csv("/content/real_distances_30_customers.csv", header = None)
df
```

```python
nodes = range(df.shape[0])
nodes
```

```python
d = [[df[j][i] for j in nodes] for i in nodes]
```

```python
origin = 0
tour_length = 0
```

```python
tour = [origin]
current_node = origin
```

```python
# 1/İNSERTİON ALGORİTHM ) Find the closest node to the origin to start the initial tour
nearest_node = None
dist_to_nearest_node = float('inf')
for i in nodes:
    if i != origin and d[origin][i] < dist_to_nearest_node:
        nearest_node = i
        dist_to_nearest_node = d[origin][i]

tour.append(nearest_node)
tour_length += dist_to_nearest_node
current_node = nearest_node

# Insertion heuristic algorithm
while len(tour) < len(nodes):
    best_increase = float('inf')
    best_node = None
    best_position = None

    # Find the best node to insert and the best position to insert it
    for node in nodes:
        if node not in tour:
            for i in range(1, len(tour)):
                prev_node = tour[i - 1]
                next_node = tour[i]
                increase = d[prev_node][node] + d[node][next_node] - d[prev_node][next_node]
                if increase < best_increase:
                    best_increase = increase
                    best_node = node
                    best_position = i

    # Insert the best node at the best position
    tour.insert(best_position, best_node)
    tour_length += best_increase
    print(f"Inserted node {best_node} at position {best_position} with increase {best_increase}")
    # Complete the tour by returning to the origin
tour.append(origin)
tour_length += d[tour[-2]][origin]

```

```python
# 2 / 2-OPT ALGORİTHM
while solution_improved:
    print()
    print('Attempting to improve the tour', current_tour,
          'with length', current_tour_length)
    solution_improved = False

    for i in range(1, len(current_tour)-2):
        for j in range(i+1, len(current_tour)-1):
            difference = round((d[current_tour[i-1]][current_tour[j]]
                              + d[current_tour[i]][current_tour[j+1]]
                              - d[current_tour[i-1]][current_tour[i]]
                              - d[current_tour[j]][current_tour[j+1]]), 2)

            print('Cost difference due to swapping', current_tour[i], 'and',
                  current_tour[j], 'is:', difference)

            if current_tour_length + difference < best_tour_length:
                print('Found an improving move! Updating the best tour...')

                best_tour = current_tour[:i] + list(reversed(current_tour[i:j+1])) + current_tour[j+1:]
                best_tour_length = round(current_tour_length + difference, 2)

                print('Improved tour is:', best_tour, 'with length',
                      best_tour_length)

                solution_improved = True

    current_tour, current_tour_length = best_tour, best_tour_length
```

```python
# 3/ 2-Opt GREEDY
def calculate_tour_length(tour, distance_matrix):
    length = 0
    for i in range(len(tour) - 1):
        length += distance_matrix[tour[i]][tour[i + 1]]
    length += distance_matrix[tour[-1]][tour[0]]  # Return to the origin
    return length

def two_opt_greedy(tour, distance_matrix):
    current_tour = tour[:]
    best_tour_length = calculate_tour_length(current_tour, distance_matrix)
    solution_improved = True

    while solution_improved:
        solution_improved = False
        for i in range(1, len(current_tour) - 2):
            for j in range(i + 1, len(current_tour) - 1):
                if j - i == 1:  # Skip adjacent nodes
                    continue
                # Calculate the difference if we swap edges (i-1, i) and (j, j+1)
                difference = (distance_matrix[current_tour[i - 1]][current_tour[j]] +
                              distance_matrix[current_tour[i]][current_tour[j + 1]] -
                              distance_matrix[current_tour[i - 1]][current_tour[i]] -
                              distance_matrix[current_tour[j]][current_tour[j + 1]])

                if best_tour_length + difference < best_tour_length:
                    # Perform the 2-opt swap
                    current_tour[i:j + 1] = reversed(current_tour[i:j + 1])
                    best_tour_length += difference
                    solution_improved = True
                    break  # Exit the inner loop
            if solution_improved:
                break  # Exit the outer loop
        if solution_improved:
            print(f"Current tour: {current_tour} with length {best_tour_length}")

    return current_tour, best_tour_length



improved_tour, improved_tour_length = two_opt_greedy(tour, d)
print("Final improved tour:", improved_tour)
print("Final improved tour length:", improved_tour_length)
```

```python
tsp = Model("tsp_model")
```

```python
x = tsp.addVars(nodes, nodes,lb = 0,vtype = GRB.BINARY, name = 'x') #x(i,j) olduğu için
u = tsp.addVars(nodes, lb = 0,vtype = GRB.INTEGER, name = 'u') #u değişkenleri sub
```

```python
tsp.addConstrs(((quicksum(x[i,j] for j in nodes if j != i) == 1) for i in nodes), name="degree_const_1") #how many nodes outgoing from i(depo hariç)
tsp.addConstrs(((quicksum(x[i,j] for i in nodes if i != j) == 1) for j in nodes), name="degree_const_2")
tsp.addConstrs(((u[i] - u[j] + len(nodes)*x[i,j] <= len(nodes) - 1) for i in nodes for j in nodes if i!=0 and j!=0),name="abc")
```

```python
tsp.setObjective((quicksum(d[i][j]*x[i,j] for i in nodes for j in nodes if i != j)),GRB.MINIMIZE)
```

```python
tsp.update()
```

```python
tsp.optimize()
```
